```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
print(lst)
data = pd.DataFrame({'whoAmI':lst})

data
```

```python
pd.get_dummies(data['whoAmI'])
```

```python

def one_hot_encode(column):
    unique_values = column.unique()
    one_hot_data = pd.DataFrame()

    for value in unique_values:
      
        print (value)
        one_hot_data[value] = (column == value).astype(int)

    return one_hot_data

one_hot_data = one_hot_encode(data['whoAmI'])

one_hot_data.head(len(lst))
```
